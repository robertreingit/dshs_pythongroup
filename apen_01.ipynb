```python
# generate n players
n = 5
x = np.arange(0.1,20*np.pi, 0.1)
players = np.zeros((x.shape[0],5))
for i in range(n):
    freq = 1.0 / (float(i+0.1) * 1.2)
    phase = np.pi/7.0 + float(i)
    players[:,i] = np.sin(x * freq + phase)
# calculate centroid
centroid = np.mean(players, axis=1)
# calculate apen for centroid
apen_centroid = ApEn(centroid, m = 1000, r = 2)

fig,ax = plt.subplots(2,1)
for i in range(n):
    ax[0].plot(x, players[:,i])
ax[1].plot(x, centroid)
ax[1].set_title('ApEn: {0}'.format(apen_centroid))
```

```python
n = 10
players = np.random.randn(10000,n)
# calculate centroid
centroid = np.mean(players, axis=1)
# calculate apen for centroid
apen_centroid = ApEn(centroid, m = 50, r = 2)

fig,ax = plt.subplots(2,1)
for i in range(n):
    ax[0].plot(players[:,i],':')
ax[1].plot(centroid)
ax[1].set_title('ApEn: {0}'.format(apen_centroid))
```
